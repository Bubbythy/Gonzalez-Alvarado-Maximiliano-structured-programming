```python
def show_menu():
    print("\n=== STREAMFLIX ===")
    print("Selecciona un género:")
    print("1. Tecnología")
    print("2. Acción")
    print("3. Ciencia Ficción")
    print("4. Drama")

def get_genre():
    option = input("\nOpción: ")
    return option

def recommend_content(genre):
    recommendations = {
        "1": ["Mr. Robot", "Silicon Valley", "Black Mirror"],
        "2": ["Deadly Code", "Operation Delta", "Chain Reaction"],
        "3": ["Altered Carbon", "Dark", "The Expanse"],
        "4": ["Breaking Bad", "Succession", "Ozark"]
    }

    if genre in recommendations:
        print("\nRecommendations:")
        for title in recommendations[genre]:
            print(f"  - {title}")
    else:
        print("\nInvalid option. Please try again.")

def rate_content():
    print("\nRate the recommended content:")
    while True:
        score = input("Enter a score from 1 to 5: ")
        if score in ["1", "2", "3", "4", "5"]:
            print(f"Thanks for rating! You gave it a {score}/5 ⭐")
            break
        else:
            print("Invalid score. Please enter a number between 1 and 5.")

def main():
    while True:
        show_menu()
        genre = get_genre()
        recommend_content(genre)
        rate_content()

        again = input("\nWould you like to perform another search? (Y/N): ").strip().upper()
        if again != "Y":
            print("\nGoodbye! Enjoy your content.")
            break

main()
```
